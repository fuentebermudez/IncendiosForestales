```python
estaciones=pd.read_json("../data/estaciones_meteorologicas.txt",encoding='UTF-8')
```

```python
estaciones['NS']=estaciones.latitud.str[-1]
estaciones['EO']=estaciones.longitud.str[-1]
```

```python
estaciones['Lat_normalizada']=estaciones.latitud.apply(lambda x: int(x[:-1])/10000)
estaciones['Lon_normalizada']=estaciones.longitud.apply(lambda x: int(x[:-1])/10000)
```

```python
def E_O(x):
    if x[12]=='W':
        x[10]=x[10]*-1
    
    return x
estaciones_transformado=estaciones.apply(E_O,axis=1)
```

```python
estaciones_transformado.to_csv("../data/estaciones_meteorologicas_normalizada.txt",encoding='UTF-8')
```
